```python
import numpy as np 
import matplotlib.pyplot as plt
```

- apple width  6 cm
- apple length  6 cm

- banana width  4 cm
- banana length  8 cm

- watermelon width  20 cm
- watermelon length  20 cm

```python
n = 100

apple_width = np.random.normal (6 , 1 , n)
apple_length = np.random.normal (6 , 1 , n)

banana_width = np.random.normal (4 , 1 , n)
banana_length = np.random.normal (8 , 1 , n)

watermelon_width = np.random.normal (20 , 1 , n)
watermelon_length = np.random.normal (20 , 1 , n)
```

- apple => red
- banana => yellow
- watermelon => green

```python
plt.scatter (apple_width , apple_length , c = 'red' , marker = '.')
plt.scatter (banana_width , banana_length , c = 'gold' , marker = '.')
plt.scatter (watermelon_width , watermelon_length , c = 'green' , marker = '.')

plt.title ('Apple , Banana & Watermelon')
plt.xlabel ('width')
plt.ylabel ('length')

plt.savefig ('outputs/apple, banana and watermelon.png')
plt.show ()
```

```python
width = np.concatenate ((apple_width , banana_width , watermelon_width))
length = np.concatenate ((apple_length , banana_length , watermelon_length))

X_train = np.array ([width , length]).T
X_train.shape
```

- Apple => 0
- Banana => 1
- Watermelon => 2

```python
y_apple = np.zeros (n , dtype = 'int')
y_banana = np.ones (n , dtype = 'int')
y_watermelon = np.ones (n , dtype = 'int') * 2

Y_train = np.concatenate ((y_apple , y_banana , y_watermelon))
Y_train.shape
```

a function for measuring distances :

```python
def euclidean_distance (x , y) :
    return np.sqrt (np.sum (x - y) ** 2)
```

Unknown fruit with following measures :
- width = 7
- Length = 9

```python
new_fruit = np.array ([7 , 9])
distances = []

for x in X_train :
    d = euclidean_distance (new_fruit , x)
    distances.append (d)

k = 7
nearest_neighbors = np.argsort (distances)[:k]

result = np.bincount (Y_train[nearest_neighbors])

if np.argmax (result) == 0 :
    print ('Apple')

elif np.argmax (result) == 1 :
    print ('Banana')

else :
    print ('Watermelon')
```
