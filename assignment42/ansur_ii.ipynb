Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from  matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from knn import KNN
```

Make Database

```python
female_data = pd.read_csv ('inputs\ANSUR II FEMALE Public.csv')
female_data = female_data.rename (columns = {'SubjectId' : 'subjectid'})
female_data.head (3)
```

```python
male_data = pd.read_csv ('inputs\ANSUR II MALE Public.csv' , encoding = 'latin_1')
male_data.head (3)
```

```python
data = pd.concat ([female_data , male_data])
data.head (3)
```

Converting unit of weight, unit of height and datatype of gender 
- female = 0
- male = 1

```python
data ['weightkg'] = data ['weightkg'] / 10

data ['stature'] = data ['stature'] / 10

data ['Gender'] = data ['Gender'].replace(['Female' , 'Male'] , [0 , 1])
```

Showing weight and stature on a scatter diagram
- female = pink
- male = blue

```python
cmap = ['pink' , 'lightblue']
plt.scatter (data['stature'] , data['weightkg'] , c = data['Gender'] , cmap= ListedColormap (cmap) , alpha=0.3)

plt.xlabel ('stature')
plt.ylabel ('weight')
plt.title ('Men and Women weight and stature')

plt.savefig ('outputs\weight&stature.png')
plt.show ()
```

- Why is the data of men higher than the data of women?
<br>Because men are heavier than women.

- Why is the data of men more right than the data of women?
<br>Because men are taller than women
